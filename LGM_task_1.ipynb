```python
#importing the library
from google.colab.patches import cv2_imshow
import cv2
```

```python
#Adding the image
img = cv2.imread("/content/leo_messi.jpg")

scale = 20 
width = int(img.shape[1] * scale / 100)
height = int(img.shape[0] * scale / 100)
dimension = (width, height)
  
# resize image
img2 = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)
cv2_imshow(img2)
```

```python
#converting to gray scale image
img2_in_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(img2_in_gray)
```

```python
#inverting the grayscale image or transforming into negative using bitwise_not operator
img2_invert = 255-img2_in_gray
cv2_imshow(img2_invert)
```

```python
#blurring the image using gaussian function
blurred_img2 = cv2.GaussianBlur(img2_invert, (21, 21), 0)
cv2_imshow(blurred_img2)
```

```python
#inverting the blurred image
blurred_invert = 255-blurred_img2
cv2_imshow(blurred_invert)
```

```python
#final sketch
final_sketch = cv2.divide(img2_in_gray, blurred_invert, scale=240.0)
cv2_imshow(final_sketch)
```
